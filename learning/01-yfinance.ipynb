```python
import yfinance as yf
```

```python
def get_data(ticker,period,interval):
    return yf.Ticker(ticker).history(period=period,interval=interval)
```

```python
def save_data(ticker,period="6mo",interval="1d"):
    df = get_data(ticker,period,interval)

    if df.empty:
        print("No data found for ticker "+ ticker)
        return

    # os.makedir(name="data/row",exist_ok=True)
    # path = f"data/row/raw/{ticker}.csv"
    # df.to_csv(path)

    # print("data saved successfully to "+ path)

    print(df.head())
```
