## Groupby operations

```python
df['Quater'] = ['Q1','Q1','Q1','Q1','Q2','Q2','Q3','Q3','Q3']
df
```

```python
groups = df.groupby('Quater')
groups.size()
```

```python
groups.mean()
```

```python
groups.max()
```

```python
groups.aggregate(['min', 'max']).round(2)
```

```python
df['Odd_Even'] = ['Odd','Even','Odd','Even','Odd','Even','Odd','Even','Odd']
df
```

```python
groups = df.groupby(['Quater','Odd_Even'])
groups.size()
```

```python
groups[['No1', 'No4']].aggregate(['sum',np.mean])
```

## Concatenation

```python
df1 = pd.DataFrame(['100','200','300','400'], index=['a','b','c','d'], columns=['A'])
df1
```

```python
df2 = pd.DataFrame(['200','150','50'], index=['f','b','d'], columns=['B'])
df2
```

```python
# joining 
df1.join(df2)
```

```python
df2.join(df1)
```

```python
df2.join(df1 , how='right')
```

```python
df2.join(df1 , how='inner')
```

```python
df2.join(df1 , how='outer')
```

```python
df = pd.DataFrame()
```

```python
df['A'] = df1
df
```

```python
df['B'] = df2
df
```

```python
df = pd.DataFrame({'A': df1['A'], 'B': df2['B']})
df
```

## Merging

```python
c = pd.Series([250, 150, 50], index=['b','d','c'])
df1['C'] = c
df2['C'] = c
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1, df2)
```

```python
pd.merge(df1, df2, on='C')
```

```python
pd.merge(df1, df2, how='outer')
```

```python
pd.merge(df1, df2, left_on='A', right_on='B')
```

```python
pd.merge(df1, df2, left_on='A', right_on='B', how='outer')
```

## Performance

```python
data = np.random.standard_normal((1000000, 2))
data.nbytes
```

```python
df = pd.DataFrame(data, columns=['X', 'y'])
df.info()
```

```python
%time res = df['X'] + df['y']
```

```python
res[:3]
```

```python
%time res = df.sum(axis=1)
```

```python
res[:3]
```

```python
%time res = df.values.sum(axis=1)
```

```python
res[:3]
```

```python
%time res = np.sum(df, axis=1)
```

```python
res[:3]
```

```python
%time res = np.sum(df.values, axis=1)
```

```python
res[:3]
```

```python
%time res = df.eval('X + y')
```

```python
res[:3]
```

```python
%time res = df.apply(lambda row: row['X'] + row['y'], axis=1)
```

```python
res[:3]
```
