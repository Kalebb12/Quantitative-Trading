## Speed comparison

```python
# pure python
import random
I = 5000

# Generating an array of 5,000 x 5,000
%time mat = [[random.gauss(0, 1) for j in range(I)] for i in range(I)]
```

```python
mat[0][:5]
```

```python
%time sum([sum(l) for l in mat])
```

```python
import sys
sum([sys.getsizeof(l) for l in mat])
```

```python
# Numpy
%time mat = np.random.standard_normal((I, I))
```

```python
%time mat.sum()
```

```python
mat.nbytes
```

```python
sys.getsizeof(mat)
```
