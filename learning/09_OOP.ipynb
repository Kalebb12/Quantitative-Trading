## **Object Oriented Programming**

```python
class HumanBeing(object):
    def __init__(self, first_name, eye_color):
        self.first_name = first_name
        self.eye_color = eye_color
        self.position = 0
    def walk_steps(self, steps):
        self.position += steps
```

```python
sandra = HumanBeing("Sandra", "black")
sandra.first_name
```

```python
sandra.eye_color
```

```python
sandra.walk_steps(10)
```

```python
sandra.position
```

```python
class FinancialInstrument(object):
    def __init__(self, symbol, price):
        self.symbol = symbol
        self.__price = price # private property
    def get_price(self):
        return self.__price
    def set_price(self, price):
        self.__price = price
```

```python
aapl = FinancialInstrument('AAPL', 100)
```

```python
aapl.get_price()
```

```python
aapl.set_price(200)
```

```python
aapl.get_price()
```

### Data model

```python
class Vector(object):
    def __init__(self, x=0, y=0, z=0):
        self.x = x
        self.y = y
        self.z = z
```

```python
v1 = Vector(1, 2, 3)
v1
```

```python
class Vector(Vector):
    def __repr__(self):
        return f"Vector({self.x}, {self.y}, {self.z})"
```

```python
v2 = Vector(1, 2, 3)
v2
```

```python
print(v2)
```

```python
class Vector(Vector):
    def __abs__(self):
        return (self.x ** 2 + self.y ** 2 + self.z ** 2) ** 0.5

    def __bool__(self):
        return bool(abs(self))
```

```python
v3 = Vector(1, 2, -1)
```

```python
abs(v3)
```

```python
bool(v3)
```

```python
v4 = Vector()
```

```python
v4
```

```python
abs(v4)
```

```python
bool(v4)
```

```python
class Vector(Vector):
    def __add__(self, other):
        x = self.x + other.x
        y = self.y + other.y
        z = self.z + other.z
        return Vector(x, y, z)

    def __mul__(self, scaler):
        return Vector(self.x * scaler, self.y * scaler, self.z * scaler)
```

```python
v5 = Vector(1, 2, 3)
```

```python
v5 + Vector(2, 3, 4)
```

```python
v5 * 3
```

```python
class Vector(Vector):
    def __len__(self):
        return 3

    def __getitem__(self, i):
        if i in [0, -3]: return self.x
        elif i in [1, -2]: return self.y
        if i in [2, -1]: return self.z
        else: raise IndexError('Index out of range.')
```

```python
v6 = Vector(1, 2, 3)
```

```python
len(v6)
```

```python
v6[0]
```

```python
v6[2]
```

```python
v6[-2]
```

```python
v6[3]
```

```python
class Vector(Vector):
    def __iter__(self):
        for i in range(len(self)):
            yield self[i]
```

```python
v7 = Vector(1,2,3)
```

```python
for i in range(3):
    print(v7[i])
```

```python
for coordinate in v7:
    print(coordinate)
```
