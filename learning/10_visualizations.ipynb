## One-dimensional data set

```python
import numpy as np
```

```python
np.random.seed(1000)
```

```python
y = np.random.standard_normal(20)
x = np.arange(len(y))
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(y.cumsum())
plt.grid()
plt.show()
```

```python
plt.plot(y.cumsum())
plt.xlim(-1,20)
plt.ylim(np.min(y.cumsum() -1), np.max(y.cumsum() + 1))
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(y.cumsum(), 'b', lw=1.5)
plt.plot(y.cumsum(), 'rx')
plt.xlabel('index')
plt.ylabel('value')
plt.title('A simple plot')
plt.show()
```

## Two-dimensional dataset

```python
y = np.random.standard_normal((20, 2)).cumsum(axis=0)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(y, 'ro')
plt.plot(y, lw=1.5)
plt.xlabel('index')
plt.ylabel('value')
plt.title('A simple plot')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(y[:, 0], lw=1.5, label='1st')
plt.plot(y[:, 1], lw=1.5, label='2nd', color='lightgreen')
plt.plot(y , 'ro')
plt.xlabel('index')
plt.ylabel('value')
plt.title('A simple plot')
plt.legend(loc=0)
plt.show()
```

```python
y[:, 0] = y[:, 0] * 100 # Scaling 1st dataset
```

```python
plt.figure(figsize=(10, 6))
plt.plot(y[:, 0], lw=1.5, label='1st')
plt.plot(y[:, 1], lw=1.5, label='2nd')
plt.plot(y, 'ro')
plt.legend(loc=0)
plt.xlabel('index')
plt.ylabel('value')
plt.title('Simple plot')
plt.show()
```

```python
fig, axl = plt.subplots() # allows the creation of a second subplot that shares the same x-axis with the first subplot
plt.plot(y[:, 0], 'b', lw=1.5, label='1st')
plt.plot(y[:, 0], 'ro')
plt.legend(loc=8)
plt.xlabel('index')
plt.ylabel('value 1st')
plt.title('A simple plot')
ax2 = axl.twinx() # shares the x axis
plt.plot(y[:,1], 'g', lw=1.5, label='2nd')
plt.plot(y[:,1], 'ro')
plt.legend(loc=0)
plt.ylabel('value 2nd')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.subplot(211) # 2 rows 1 column and index position 1
plt.plot(y[:,0], lw=1.5, label='1st')
plt.plot(y[:,0], 'ro')
plt.legend(loc=0)
plt.ylabel('value')
plt.title('A simple plot')
plt.subplot(212) # 2 rows 1 column and index position 2
plt.plot(y[:,1], 'g', lw=1.5, label='2nd')
plt.plot(y[:,1], 'ro')
plt.legend(loc=0)
plt.ylabel('value')
plt.xlabel('index')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.subplot(121)
plt.plot(y[:, 0], lw=1.5, label='1st')
plt.plot(y[:, 0], 'ro')
plt.legend(loc=0)
plt.xlabel('index')
plt.ylabel('values')
plt.title('1st data set')
plt.subplot(122)
plt.bar(np.arange(len(y)), y[:, 1], width=0.5, color='g', label='2nd')
plt.legend(loc=0)
plt.xlabel('index')
plt.title('2nd data set')
plt.show()
```
