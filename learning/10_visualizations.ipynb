## One-dimensional data set

```python
import numpy as np
```

```python
np.random.seed(1000)
```

```python
y = np.random.standard_normal(20)
x = np.arange(len(y))
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(y.cumsum())
plt.grid()
plt.show()
```

```python
plt.plot(y.cumsum())
plt.xlim(-1,20)
plt.ylim(np.min(y.cumsum() -1), np.max(y.cumsum() + 1))
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(y.cumsum(), 'b', lw=1.5)
plt.plot(y.cumsum(), 'rx')
plt.xlabel('index')
plt.ylabel('value')
plt.title('A simple plot')
plt.show()
```

## Two-dimensional dataset

```python
y = np.random.standard_normal((20, 2)).cumsum(axis=0)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(y, 'ro')
plt.plot(y, lw=1.5)
plt.xlabel('index')
plt.ylabel('value')
plt.title('A simple plot')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(y[:, 0], lw=1.5, label='1st')
plt.plot(y[:, 1], lw=1.5, label='2nd', color='lightgreen')
plt.plot(y , 'ro')
plt.xlabel('index')
plt.ylabel('value')
plt.title('A simple plot')
plt.legend(loc=0)
plt.show()
```

```python
y[:, 0] = y[:, 0] * 100 # Scaling 1st dataset
```

```python
plt.figure(figsize=(10, 6))
plt.plot(y[:, 0], lw=1.5, label='1st')
plt.plot(y[:, 1], lw=1.5, label='2nd')
plt.plot(y, 'ro')
plt.legend(loc=0)
plt.xlabel('index')
plt.ylabel('value')
plt.title('Simple plot')
plt.show()
```

```python
fig, axl = plt.subplots() # allows the creation of a second subplot that shares the same x-axis with the first subplot
plt.plot(y[:, 0], 'b', lw=1.5, label='1st')
plt.plot(y[:, 0], 'ro')
plt.legend(loc=8)
plt.xlabel('index')
plt.ylabel('value 1st')
plt.title('A simple plot')
ax2 = axl.twinx() # shares the x axis
plt.plot(y[:,1], 'g', lw=1.5, label='2nd')
plt.plot(y[:,1], 'ro')
plt.legend(loc=0)
plt.ylabel('value 2nd')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.subplot(211) # 2 rows 1 column and index position 1
plt.plot(y[:,0], lw=1.5, label='1st')
plt.plot(y[:,0], 'ro')
plt.legend(loc=0)
plt.ylabel('value')
plt.title('A simple plot')
plt.subplot(212) # 2 rows 1 column and index position 2
plt.plot(y[:,1], 'g', lw=1.5, label='2nd')
plt.plot(y[:,1], 'ro')
plt.legend(loc=0)
plt.ylabel('value')
plt.xlabel('index')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.subplot(121)
plt.plot(y[:, 0], lw=1.5, label='1st')
plt.plot(y[:, 0], 'ro')
plt.legend(loc=0)
plt.xlabel('index')
plt.ylabel('values')
plt.title('1st data set')
plt.subplot(122)
plt.bar(np.arange(len(y)), y[:, 1], width=0.5, color='g', label='2nd')
plt.legend(loc=0)
plt.xlabel('index')
plt.title('2nd data set')
plt.show()
```

```python
y = np.random.standard_normal((1000, 2))
```

```python
plt.figure(figsize=(10, 6))
plt.plot(y[:,0], y[:,1], 'ro')
plt.xlabel('1st')
plt.ylabel('2nd')
plt.title('Scatter plot')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(y, label=['1st', '2nd'], bins=25, color=['g','m'])
plt.legend(loc=0)
plt.xlabel('value')
plt.ylabel('frequency')
plt.title('Histogram')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(y, label=['1st', '2nd'], bins=20, color=['g','m'], alpha=0.5, stacked=True)
plt.legend(loc=0)
plt.xlabel('value')
plt.ylabel('frequency')
plt.title('Histogram')
plt.show()
```

```python
fig,axl = plt.subplots(figsize=(10, 6))
plt.boxplot(y)
plt.setp(axl, xticklabels=['1st', '2nd'])
plt.xlabel('data set')
plt.ylabel('value')
plt.title('Boxplot')
plt.show()
```

```python
def func(x):
    return 0.5 * np.exp(x) + 1
a,b = 0.5, 1.5
x = np.linspace(0, 2)
y = func(x)
Ix = np.linspace(a, b)
Iy = func(Ix)
verts =[(a, 0)] + list(zip(Ix, Iy)) + [(b, 0)]
```

```python
from matplotlib.patches import Polygon
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
plt.plot(x, y, 'b', lw=2)
plt.ylim(ymin=0)
poly = Polygon(verts, facecolor='0.7', edgecolor='0.5')
ax.add_patch(poly)
plt.text(0.5 * (a + b), 1,r'$\int_a^b f(x)\mathrm{d}x$', horizontalalignment='center', fontsize=20)
plt.figtext(0.9, 0.075, '$x$')
plt.figtext(0.075, 0.9, '$f(x)$')
ax.set_xticks((a, b))
ax.set_xticklabels(('$a$', '$b$'))
plt.show()
```
