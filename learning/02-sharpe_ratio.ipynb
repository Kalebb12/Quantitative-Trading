## Calculating sharp ratio

```python
import yfinance as yf
```

```python
avgo = yf.download("AVGO",period="1y")["Close"].pct_change().dropna()
```

```python
avgo.head()
```

### Standard deviation (Risk)

```python
a = avgo.std() * (252**0.5)
```

```python
b = avgo.cumsum().iloc[-1]
```

## Returns

```python
b/a
```
